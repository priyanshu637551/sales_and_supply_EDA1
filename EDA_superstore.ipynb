## Univariate Analysis

```python
print("\n=== CATEGORIAL VARIABLE ===\n")

categorical_cols = ['Customer Segment','Category','Region','Payment Mode','Delivery Status','Auto Reorder']

for col in categorical_cols:
    print(f'\n --- {col}---')
    print(df[col].value_counts())

    #visualization
    plt.figure(figsize = (6,6))
    df[col].value_counts().plot(kind = 'bar' , color = 'pink')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation = 0)
    plt.tight_layout()
    plt.show()

```

## Birvariate analysis
Birvarit analysis is the statical analysis of two variable at the same tine . The core purpose is to investigate the relationship between them.

```python
print('\n=== sales & profit by category ===\n')
category_performance = df.groupby('Category').agg({
    'Sales Amount' : 'sum',
    'Profit':'sum',
    'Order ID': 'count'
}).rename(columns = {'Order ID': 'Order Count'})

category_performance['Profit Margin'] =round((category_performance['Profit']/category_performance['Sales Amount'])*100,2)
print(category_performance , '\n\n')

#visualization
fig , axes =  plt.subplots(2,2,figsize=(15,12))

#sales category
category_performance['Sales Amount'].plot(kind= 'bar',ax=axes[0,0],color = 'skyblue',alpha = 0.7)
axes[0,0].set_title('total sales by category')
axes[0,0].tick_params(axis = 'x',rotation = 30)

#profit category
category_performance['Profit'].plot(kind= 'bar',ax=axes[0,1],color = 'orange',alpha = 0.7)
axes[0,1].set_title('total profit by category')
axes[0,1].tick_params(axis = 'x',rotation = 30)

#profit margin category
category_performance['Profit Margin'].plot(kind= 'bar',ax=axes[1,0],color = 'lightgreen',alpha = 0.7)
axes[1,0].set_title('total Profit Margin by category')
axes[1,0].tick_params(axis = 'x',rotation = 30)

#order category
category_performance['Order Count'].plot(kind= 'bar',ax=axes[1,1],color = 'yellow',alpha = 0.7)
axes[1,1].set_title('Order Count by category')
axes[1,1].tick_params(axis = 'x',rotation = 30)

plt.tight_layout()
plt.show()
```

## Regional Performance 

```python
print('\n=== REGIONAL PERFORMANCE ===\n')
regional_performance = df.groupby('Category').agg({
    'Sales Amount' : 'sum',
    'Profit':'sum',
    'Order ID': 'count'
}).rename(columns = {'Order ID': 'Order Count'})

regional_performance['Profit Margin'] =round((regional_performance['Profit']/regional_performance['Sales Amount'])*100,2)
print(regional_performance,'\n')

#visualization
plt.figure(figsize = (12,8))

plt.subplot(2,2,1)
regional_performance['Sales Amount'].plot(kind = 'pie' , autopct='%1.1f%%')
plt.title('Sales Distribution by region')

plt.subplot(2,2,2)
regional_performance['Profit'].plot(kind = 'pie' , autopct='%1.1f%%')
plt.title('Profit Distribution by region')

plt.subplot(2,2,3)
regional_performance['Profit Margin'].plot(kind = 'pie' , autopct='%1.1f%%')
plt.title('Profit Margin Distribution by region')

plt.subplot(2,2,4)
regional_performance['Order Count'].plot(kind = 'pie' , autopct='%1.1f%%')
plt.title('Order Count Distribution by region')


plt.tight_layout()
plt.show()
```

### TIME SERIES ANALYSIS

```python
print('\n=== time series analysis ===\n')

#time componenets
df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Order Year'] = df['Order Date'].dt.year

#monthly trends

monthly_trends= df.groupby('Order Month').agg({
    'Sales Amount' : 'sum',
    'Profit':'sum',
    'Order ID': 'count'
}).rename(columns = {'Order ID': 'Order Count'})

print('monthly trends:')
print(monthly_trends.tail(10),'\n\n')


#visualization
plt.figure(figsize = (15,10))

plt.subplot(3,1,1)
monthly_trends['Sales Amount'].plot(kind = 'line' , color = 'skyblue' , marker = 'o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales Amount')
plt.grid(True)

plt.subplot(3,1,2)
monthly_trends['Profit'].plot(kind = 'line' , color = 'red' , marker = 'o')
plt.title('Monthly Profit Trend')
plt.ylabel('Profit')
plt.grid(True)


plt.subplot(3,1,3)
monthly_trends['Order Count'].plot(kind = 'line' , color = 'lightgreen' , marker = 'o')
plt.title('Monthly Order Count Trend')
plt.ylabel('Order Count')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Advance Analysis

```python
print('\n ===  Advance Analysis ===\n')

#select numerical columns for correlation
numerical_df = df[['Quantity', 'Unit Price','Discount %','Sales Amount','Cost Price','Profit']]

# correlation matrix
correlation_matrix = numerical_df.corr()
print('correlation matrix:')
print(correlation_matrix , '\n\n')

#heatmap
plt.figure(figsize = (12,10))
sns.heatmap(correlation_matrix, annot = True , cmap = 'coolwarm' , center = 0,square = True, linewidth = 0.5)
plt.title('correlation_matrix')
plt.tight_layout()
plt.show()
```

### Customer Segment Analysis

```python
print('\n=== customer segment Analysis ===\n')


segment_analysis= df.groupby('Customer Segment').agg({
    'Sales Amount' : ['sum','mean'],
    'Profit':['sum','mean'],
    'Order ID': 'count',
    'Discount %':'mean'
}).round(2)

segment_analysis.columns = ['Total Sales','Avg Sales','Total Profit','Avg Profit','Order Count','Avg Discount']
print(segment_analysis,'\n\n')

#visualization
fig , axes =  plt.subplots(2,2,figsize=(15,10))

#sales category
segment_analysis['Total Sales'].plot(kind= 'bar',ax=axes[0,0],color = 'skyblue',alpha = 0.7)
axes[0,0].set_title('total sales by category')
axes[0,0].tick_params(axis = 'x',rotation = 30)

#profit category
segment_analysis['Total Profit'].plot(kind= 'bar',ax=axes[0,1],color = 'orange',alpha = 0.7)
axes[0,1].set_title('total profit by category')
axes[0,1].tick_params(axis = 'x',rotation = 30)

#profit margin category
segment_analysis['Avg Profit'].plot(kind= 'bar',ax=axes[1,0],color = 'lightgreen',alpha = 0.7)
axes[1,0].set_title('Avg Profit Margin by category')
axes[1,0].tick_params(axis = 'x',rotation = 30)

#order category
segment_analysis['Avg Discount'].plot(kind= 'bar',ax=axes[1,1],color = 'yellow',alpha = 0.7)
axes[1,1].set_title('Avg Discount by category')
axes[1,1].tick_params(axis = 'x',rotation = 30)

plt.tight_layout()
plt.show()
```

## Summary

```python
print('\n ===  Key Insight and Summary === \n')

# overall Matrices
total_sales = df['Sales Amount'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
avg_profit_margin = (total_profit / total_sales) * 100

print(f'Overall performance :\n')
print(f'total sales : ${total_sales:,.2f}')
print(f'total profit : ${total_profit:,.2f}')
print(f'total orders {total_orders}')
print(f'Average Profit margin : {avg_profit_margin:.2f}%')

# top categories by profit
top_categories = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(f'\nTop categories:\n')
for i, (category, profit) in enumerate(top_categories.items(), 1):   # start=1
    print(f'{i}. {category} : ${profit:,.2f}')

# regional Performance - pick the metric you want, e.g. Sales Amount
best_region_by_sales = df.groupby('Region')['Sales Amount'].sum().idxmax()
print(f'\nBest Performing region by Sales : {best_region_by_sales}')

# or best region by Profit
best_region_by_profit = df.groupby('Region')['Profit'].sum().idxmax()
print(f'Best Performing region by Profit : {best_region_by_profit}')

# delivery Performance
delivery_stats = df['Delivery Status'].value_counts(normalize=True) * 100
print(f'\nDelivery Status (%):')
for status, percentage in delivery_stats.items():
    print(f'{status}: {percentage:.1f}%')

# customer segment (make sure column name matches dataframe)
# If your column is 'Customer Segment' use that exact name
best_segment_by_profit = df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f'\nBest Performing customer segment by Profit : {best_segment_by_profit}')
```

### Export Results

```python
analysis_results = {
    'total_sales': total_sales,
    'total_profit': total_profit,
    'total_orders': total_orders,
    'avg_profit_margin': avg_profit_margin,
    'top_category': top_categories.index[0],
    'best_region': best_region_by_profit,
    'best_customer_segment': best_segment_by_profit
}

# Convert to DataFrame and save
results_df = pd.DataFrame([analysis_results])
results_df.to_csv('eda_analysis_summary.csv', index=False)

print("\nAnalysis summary saved to 'eda_analysis_summary.csv'")
```
